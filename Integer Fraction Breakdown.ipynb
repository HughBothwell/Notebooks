# 1/7 = 1/a + 1/b + 1/c + 1/d
for 0 < a < b < c < d in Z

Problem at https://twitter.com/sonukg4india/status/1680247216714264576

Here I have implemented my posted solution; it finds 1865 unique solutions:

```python
from math import ceil, floor
import sympy as sp

# keep a list of solutions found
sols = []

# ((1/7)/4) < 1/a < 1/7
#   7 < a < 28
for a in range(8, 28):
    # 1/b < (1/7 - 1/a)
    #   7a / (a - 7) < b
    min_b = ceil((7*a + 1) / (a - 7))
    # but also a < b
    min_b = max(a+1, min_b)
    # (1/7 - 1/a)/3 < 1/b
    #   b < 21a / (a - 7)
    max_b = floor((21*a - 1) / (a - 7))
    
    print(f"a = {a}: b in {min_b}..{max_b}")    
    for b in range(min_b, max_b + 1):
        # 1/c < (1/7 - 1/a - 1/b)
        #   7ab / (ab - 7b - 7a) < c
        min_c = ceil((7*a*b + 1) / (a*b - 7*b - 7*a))
        # but also b < c
        min_c = max(b+1, min_c)
        # (1/7 - 1/a - 1/b)/2 < 1/c
        #   c < 14ab / (ab - 7b - 7a)
        max_c = floor((14*a*b - 1) / (a*b - 7*b - 7*a))
        
        print(f"  c in {min_c}..{max_c}")
        for c in range(min_c, max_c + 1):
            # 1/7 - 1/a - 1/b - 1/c = 1/d
            #   d = 7abc / (abc - 7bc - 7ac - 7ab)
            # but is only a solution if the result is an integer!
            d_num = 7*a*b*c
            d_denom = a*b*c - 7*(b*c + a*c + a*b)
            # (also it's good to not divide by 0)
            if d_denom and not d_num % d_denom:
                d = d_num // d_denom
                # and the solution is only unique if c < d
                if d > c:
                    print(f"    {a} {b} {c} {d}  <-- solution")
                    sols.append((a,b,c,d))
                    
print(f"{len(sols)} solutions found")
```
