```python
import numpy as np
import matplotlib.pyplot as plt

MAX_DIST = 20.
MIN_DIST = 0.5

def w1(d):
    return (d ** 2. + 20 * d) ** 0.5 / (10 * d)

ds = np.linspace(MAX_DIST, MIN_DIST, 100)
ws = w1(ds)

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(ds, ws, "r-")
plt.xlim([MAX_DIST, 0.])
plt.ylim([0., 1.])

plt.show()
```
